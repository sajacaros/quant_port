먼저 샘플로 사용할 주가 및 재무제표 데이터를 다운로드 받는다. 데이터는 아래 링크에 .sql 파일로 업로드 되어 있다.

https://drive.google.com/file/d/13KLFlZTGJvyrlXQYQ_mR0RgAbfYtufti/view?usp=sharing



[다운로드] 버튼을 눌러 파일을 다운로드 한 후 SQL에서 해당 데이터베이스를 불러온다.



1. Navigator에서 Administration 부분을 클릭한 후 Data Import를 선택한다.
2. Import from Self-Contained File를 선택한 후 […]을 눌러 다운로드 받은 파일을 선택한다.
3. Default Target Schema 우측의 New를 누른 후 저장될 데이터베이스 이름을 입력한다.
4. 하단의 Start Import를 클릭한다.



데이터베이스를 확인해보면 **티커, 섹터, 주가, 재무제표, 밸류에이션** 데이터가 들어와 있다.
